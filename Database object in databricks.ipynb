```sql
create function if not exists full_name_udf(firstname string, lastname string)
returns string
-- return concat(first_name,' ',last_name)
return firstname || ' ' || lastname
```

```sql
select customer_id, full_name_udf(firstname,lastname) as full_name from customer_reviews
```

```sql
create function sentiments_analysis(feedback string)
returns string
return 
case
when feedback like '%Good%' then 'positive'
when feedback like '%noot good%' then 'negative'
else 'neutral'
End
```

```sql
select customer_id, feedback , sentiments_analysis(feedback) as sentiments from customer_reviews
```

```sql
create or replace view customer_sentiments as
select customer_id, feedback , sentiments_analysis(feedback) as sentiments from customer_reviews;
select * from customer_sentiments
```

```sql
create or replace view customer_sentiments as
select sentiments_analysis(feedback) as sentiments,
 count (*) as count
from customer_reviews
group by sentiments
```
