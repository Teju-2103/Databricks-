```sql
-- CSV → allows spaces in column names 
-- Delta table → does NOT allow spaces & special characters 
-- saveAsTable() creates a Delta table, so it fails

%python
from pyspark.sql.functions import col
df=spark.read.csv('/Volumes/catlog_sample/bronze/raw_csv_file',header=True,inferSchema=True)
df_clean = df.select(
    col("Customer Id").alias("customer_id"),
    col("First Name").alias("first_name"),
    col("Last Name").alias("last_name"),
    col("Phone 1").alias("phone_1"),
    col("Phone 2").alias("phone_2"),
    col("Subscription Date").alias("subscription_date")
)
df_clean.write.saveAsTable("catlog_sample.bronze.data_pyspark2")
```

```sql
CREATE TABLE catlog_sample.bronze.customertable_sql AS
SELECT
  `Customer Id`        AS customer_id,
  `First Name`         AS first_name,
  `Last Name`          AS last_name,
  `Phone 1`            AS phone_1,
  `Phone 2`            AS phone_2,
  `Subscription Date`  AS subscription_date,
  current_date()       AS ingestion_date
FROM read_files(
  '/Volumes/catlog_sample/bronze/raw_csv_file',
  format => 'csv',
  header => true
);
```
